```python
# DO FOR my computer
indices = []
with open('mining.log', 'r') as fp:
    line = fp.readline()
    while line:
        line = line.strip()
        if 'Record index:' in line:
            r1 = line.split('Record index: ')
            num_str = r1[1].split(',')[0]
            indices.append(int(num_str))

        line = fp.readline()
```

```python
# DO FOR EC2
indices2 = []
with open('ec2/mining.log', 'r') as fp:
    line = fp.readline()
    while line:
        line = line.strip()
        if 'Record index:' in line:
            r1 = line.split('Record index: ')
            num_str = r1[1].split(',')[0]
            indices2.append(int(num_str))

        line = fp.readline()
```

```python
indices.extend(indices2)
indices
```

```python
import numpy as np
from matplotlib import pyplot as plt

indices_np = np.array(indices)
ones = np.ones(indices_np.shape)
plt.scatter(indices, ones)
plt.xlim((0, 1000))
plt.show()
```
